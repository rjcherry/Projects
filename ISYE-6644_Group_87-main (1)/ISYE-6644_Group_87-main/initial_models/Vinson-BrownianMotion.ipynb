```python
from matplotlib import pyplot as plt
import numpy as np
```

# Simulating Brownian Motion

Params:
- $S_0 > 0$ initial stock price
- $\mu$ coefficients for drifts
- $\sigma$ constant volatility

Brownian motion equation:
$$dS_t = \mu S_t dt + \sigma S_t dW_t$$

Explicit form:
$$S_{t+1} = S_{t} \exp\left[\left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W \right]$$

```python
### Initial setup
# number of steps
n = 100
# time 
T = 1
# number of simulations
M = 50

### Params
# drift coefficent
mu = 0.1
# initial stock price 
S0 = 100
# volatility
sigma = 0.3
```

```python
### Brownian Motion Paths
# time step
dt = T / n

# Set up St for each step
St = np.exp((mu - np.square(sigma) / 2) * dt + sigma * np.random.normal(0, np.sqrt(dt), size=(n - 1, M)))
# First value starts at 1
St = np.vstack([np.ones(M), St])
# Calculate each step value
St = S0 * St.cumprod(axis=0)
```

```python
### Graph result
time = np.linspace(0, T, n)

# Define x values for each series
x = np.full(shape=(M, n), fill_value=time).T

# Plot results
plt.plot(x, St)
plt.xlabel("Years")
plt.ylabel("Stock price")
plt.title("Brownian Motion");
```
