```python
tictactoeboard = {'A1': ' ', 'B1':' ', 'C1': ' ',
                  'A2': ' ', 'B2':' ', 'C2': ' ',
                  'A3': ' ', 'B3':' ', 'C3': ' '}

def tictactoe(position):
    print(position['A1'] + '|' + position['B1'] + '|' + position['C1'])
    print('-+-+-')
    print(position['A2'] + '|' + position['B2'] + '|' + position['C2'])
    print('-+-+-')
    print(position['A3'] + '|' + position['B3'] + '|' + position['C3'])


def gameplay():
    
    Move = 'X'
    Counter= 0
    Wins = True

    for u in range(200):
        tictactoe(tictactoeboard)
        print("Your go," + Move + ".Move where?")

        go = input()        

        if tictactoeboard[go] == ' ':
            tictactoeboard[go] = Move
            Counter += 1
        else:
            print("Space occupied. Pick another position!")
            continue


        if Counter >= 5:
            if tictactoeboard['A1'] == tictactoeboard['B2'] == tictactoeboard['C3'] != ' ': # diagonal
                tictactoe(tictactoeboard)              
                print(Move + " wins.")
                Wins == True
                break
    
            elif tictactoeboard['A3'] == tictactoeboard['B2'] == tictactoeboard['C1'] != ' ': #diagonal2
                tictactoe(tictactoeboard)
                print(Move + " wins!")
                Wins == True
                break
        
            elif tictactoeboard['A1'] == tictactoeboard['B1'] == tictactoeboard['C1'] != ' ': #across top
                tictactoeboard(tictactoeboard)
                print(Move + " wins!")
                Wins == True
                break
    
            elif tictactoeboard['A2'] == tictactoeboard['B2'] == tictactoeboard['C2'] != ' ': #across middle
                tictactoe(tictactoeboard)
                print(Move + " wins!")
                Wins == True
                break
    
            elif tictactoeboard['A3'] == tictactoeboard['B3'] == tictactoeboard['C3'] != ' ': #across bottom
                tictactoe(tictactoeboard)
                print(Move + " wins!")
                Wins == True
                break
    
            elif tictactoeboard['A1'] == tictactoeboard['A2'] == tictactoeboard['A3'] != ' ': #down left column
                tictactoe(tictactoeboard)
                print(Move + " wins!")
                Wins == True
                break
    
            elif tictactoeboard['B1'] == tictactoeboard['B2'] == tictactoeboard['B3'] != ' ': #down middle column
                tictactoe(tictactoeboard)
                print(Move + " wins!")
                Wins == True
                break
    
            elif tictactoeboard['C1'] == tictactoeboard['C2'] == tictactoeboard['C3'] != ' ': #down right column
                tictactoe(tictactoeboard)
                print(Move + " wins!")
                Wins == True
                break
         

        if Counter == 9:
            print("It's a tie! No one wins!")
            tictactoe(tictactoeboard)
            break
            
        
        if Move == 'O':
            Move = 'X'
        else:
            Move = 'O'
            
            
gameplay()
```
